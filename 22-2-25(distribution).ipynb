```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.stats import bernoulli

p=0.6
data=bernoulli.rvs(p,size=1000)
plt.bar([0,1],[sum(data==0),sum(data==1)],color=['red','green'],tick_label=["failure(0)","Sucess(1)"])
plt.xlabel("outcome")
plt.ylabel("frequency")
plt.title("Bernoulli Distribution(p=0.6)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 70
sigma = 10

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='red', label=f"N({mu},{sigma}^2)")
plt.xlabel("Exam Score")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='blue',label="standard deviation ")
plt.xlabel("z score")
plt.ylabel("probability density")
plt.title("standard normal")
plt.axvline(x=0,color='red',linestyle='--',label="mean(z=0)")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
np.random.seed(42)
income_data=np.random.exponential(scale=50,size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data,bins=50,kde=True,color="red")
plt.title("original income distribution(skwed)")
plt.xlabel("income ($1000)")
plt.ylabel("frequency")
plt.show()



sample_means=[]
n_samples=1000
sample_size=30
for _ in range (n_samples):
    sample=np.random.choice(income_data,size=sample_size)
    sample_means.append(np.mean(sample))
    
plt.figure(figsize=(8,4))
sns.histplot(sample_means,bins=50,kde=True,color="red")
plt.xlabel("sample mean")
plt.ylabel("frequency")
plt.show()
```
